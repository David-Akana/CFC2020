```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Vocalnic eruption data

```python
volc = pd.read_csv('data/txt/volerup.txt',delimiter = '\t', quoting = 3, encoding='latin-1')
```

```python
volc.head()
```

```python
volc.isnull().sum()
```

```python
volc  = volc[['Year', 'Name', 'Location','Country','Latitude','Longitude']]
```

```python
volc.head()
```

```python
# Repition shows that there were places that the eruption happend more than once.
print(volc.Latitude.duplicated().sum())
print(volc.Longitude.duplicated().sum())
```

```python
print(volc.Name.duplicated().sum())
```
